```python
import numpy as np
from collections import Counter

#step 1: create a small dataset
# Features :[height,Weight], Labels: 0 = Short, 1 = Tall
X_train=np.array([
    [150,50],
    [160,55],
    [170,60],
    [180,65],
    [190,70]
])
Y_train=np.array([0,0,1,1,1])  # corresponding labels
```

```python
# step 2: define euclidean distance function 
def euclidian_dist(X1,X2):
    return np.sqrt(np.sum((X1-X2)**2))
```

```python
# step  3: KNN classifier
def knn_predict(X_train, Y_train, X_test, k=3):
    # compute distances to all training samples
    dist=[euclidian_dist(X_test, X_train) for X_train in X_train ]
    k_index=np.argsort(dist)[:k]
    k_nearlabls=[Y_train[i] for i in k_index]
    most_common=Counter(k_nearlabls).most_common(1)
    return most_common[0][0]
```

```python
x_new=np.array([155,70])
prediction= knn_predict(X_train,Y_train,x_new,k=3)
print("predicted class for",x_new, "is:", prediction)
```
